```python
#Focusing on Department nr. 4, which is the produce department
produce_dept_customer_profiles = departments_customer_profiles[departments_customer_profiles['department_id'] == 4]
```

```python
produce_dept_customer_profiles.head(60)
```

```python
#Creating a bar chart showing purchases from the produce department by different customer profiles
bar_chart_produce_dept_customer_profiles = produce_dept_customer_profiles[['Customer_Profile']].value_counts().plot.barh(color= 'lightblue', title='Items Purchased from the Produce Department by Customer Profile')
bar_chart_produce_dept_customer_profiles.legend(loc='center left',bbox_to_anchor=(1, 0.5))
bar_chart_produce_dept_customer_profiles.set(xlabel='Number of Items Purchased from the Produce Department', ylabel='Customer Profile', )
```
