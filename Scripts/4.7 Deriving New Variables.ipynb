## Create a new variable indicating the price-range of the items

```python
#Create a subset of ords_prods_merge to avoid memory issues
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
#Create and define a price label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Use the price_label function on the subset called df
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the values in the"price_range" column
df['price_range'].value_counts()
```

```python
#Check what the most expensive product within the subset is
df['prices'].max()
```

Use the loc() function to add a price label on the entire dataframe

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

## Create a "busiest_day" column showing the busiest and least busy day of the week

```python
#Print the frequency of the “orders_day_of_week” column to find the busiest and least busy day
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna=False)
```

```python
#Create a function to show the busiest and least busy day of the week
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Add a "busiest_day" column to the dataframe
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## Create a "busiest_days" column showing the busiest and least busy 2 days of the week

Refer to the frequency of the “orders_day_of_week” column to find:

The two busiest days: 0 and 1

The two least busy days: 3 and 4

```python
# Create a for-loop which will be used to add a "busiest_days" column
results = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    results.append("Busiest days")
  elif value ==4 or value == 3:
    results.append("Least busy days")
  else:
    results.append("Regularly busy days")
```

```python
results
```

```python
#Add a "busiest_days" column to the dataframe
df_ords_prods_merged['busiest_days'] = results
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

We can see that the number of rows for "Busiest days" corresponds to the number of rows with the value "0" or "1" in the "orders_day_of_week" column, and the number of rows for "Least busy days" corresponds to the number of rows with the value "3" or "4" in the "orders_day_of_week" column.

## Create a "busiest_period_of_day" column

```python
#Print the frequency of the “order_hour_of_day” column to find the busiest and least busy hour of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

The two hours when the most orders are made are 10 and 11. The two hours when the fewest orders are made are 4 and 3.

```python
# Create a for-loop which will be used to add a "busiest_period_of_day" column
results_period = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 10 or value == 11:
    results_period.append("Most orders")
  elif value ==4 or value == 3:
    results_period.append("Fewest orders")
  else:
    results_period.append("Average orders")
```

```python
results_period
```

```python
#Add a "busiest_period_of_day" column to the dataframe
df_ords_prods_merged['busiest_period_of_day'] = results_period
```

```python
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

We have successfully created a “busiest_period_of_day" column, with time periods labelled.
